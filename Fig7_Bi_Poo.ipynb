```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# List of points
points = [(-5,-1),(-4,1),(0,1),(15,1),(30, 1), (45, 1), (60, 1), (75, 1), (84, 1),(85,-1),(90,-1),(100,-1),(200,-1),(300,-1),(400,-1),(500,-1),(-100,-1),(-200,-1),(-300,-1),(-400,-1),(-500,-1),(-15,-1),(-30,-1),(-45,-1),(-60,-1),(-75,-1),(-90,-1)]

# Convert the list of points to numpy arrays for x and y coordinates
x_coords, y_coords = zip(*points)
colors = ['navy' if y == 1 else 'purple' if y == -1 else 'blue' for y in y_coords]

# Plot the points
plt.scatter(x_coords, y_coords,  color=colors)  # 'scatter' is used to plot points
# Draw vertical lines at x=0 and x=84
plt.axvline(x=-4, color='red', linestyle='--')  # Dashed line at x=0
plt.axvline(x=84, color='green', linestyle='--')  # Dashed line at x=84

# Draw x and y axes
plt.axhline(y=0, color='black', linewidth=0.5)  # y-axis
plt.axvline(x=0, color='black', linewidth=0.5)  # x-axis

# Denotation (labels) for vertical lines
plt.text(-4, max(y_coords) - 0.5, ' t=-4', horizontalalignment='center', color='black')
plt.text(84, max(y_coords) - 0.5, ' t=84', horizontalalignment='center', color='black')

# Annotations for LTP and LTD along the left of the plot (outside the grid)
left_boundary = min(x_coords) - 2  # Move a bit more to the left from the leftmost point for clarity
plt.text(left_boundary, 1, 'LTP', horizontalalignment='left', color='black', fontsize=15)
plt.text(left_boundary, -1, 'LTD', horizontalalignment='left', color='black', fontsize=15)

# Setting some features for clarity
plt.grid(True)  
plt.xlabel('Time(s)')
plt.ylabel('Status of LTP/D')
plt.title('Change of LTP/D status over time')
plt.legend(loc='upper right')
# Show the plot
plt.show()




```
