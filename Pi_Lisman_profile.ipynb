```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch
```

```python
def count_matrix(matrix, value):
    """
    Count the occurrences of a specific value in a matrix.

    Parameters:
    matrix (np.array): The numpy array (matrix) to search in.
    value: The value to count in the matrix.

    Returns:
    int: The number of occurrences of the value in the matrix.
    """
    # Count the occurrences of the specified value
    return np.sum(matrix == value)
```

```python
def phase_diagram(m,step1,step2):
    """
    Plots a given matrix with specific color mappings and customizations.

    Args:
    m (numpy.ndarray): A 2D numpy array representing the matrix to be plotted.

    Returns:
    None: The function creates and displays a plot.
    """
    # Define colors and colormap
    colors = ['white', 'blue', 'red']  # Order is important here
    cmap = ListedColormap(colors)

    # Convert matrix values to match the colormap
    m_copy = m.copy()
    m_copy[m_copy == 0] = 0   # Mapping 0 to white
    m_copy[m_copy == 1] = 1   # Mapping 1 to blue
    m_copy[m_copy == -1] = 2  # Mapping -1 to red

    # Display the matrix
    fig, ax = plt.subplots(figsize=(6, 5))  # Adjusted to a smaller figure size
    ax.imshow(m_copy, cmap=cmap, aspect='auto', origin='lower')

    # Display grid/lattice for the matrix
    ax.set_xticks(np.arange(-0.5, m.shape[1], 1), minor=True)
    ax.set_yticks(np.arange(-0.5, m.shape[0], 1), minor=True)
    ax.grid(which='minor', color='black', linestyle='-', linewidth=1, alpha=0.1)

    # Set custom tick labels for y-axis (T(seconds))
    y_ticks = np.arange(0, m.shape[0], 10)
    y_labels = ["{:.1f}".format(step1 * i) for i in range(10, m.shape[0]+1, 10)]
    ax.set_yticks(y_ticks)
    ax.set_yticklabels(y_labels)
    ax.set_ylabel('T(seconds)')

    # Set custom tick labels for x-axis (Calcium Impulse(μm))
    x_ticks = np.arange(4, m.shape[1], 5)
    x_labels = ["{:.1f}".format(step2 * i) for i in range(5, m.shape[1]+1, 5)]
    ax.set_xticks(x_ticks)
    ax.set_xticklabels(x_labels)
    ax.set_xlabel('Calcium Impulse(μm)')

    # Create legend
    legend_labels = {
        'red': 'LTD',
        'blue': 'LTP',
        'white': 'basal'
    }
    handles = [Patch(facecolor=color, edgecolor='black', label=label) for color, label in legend_labels.items()]
    ax.legend(handles=handles, loc='upper right')

    plt.tight_layout()
    plt.show()

    # Save the figure
    fig.savefig('phase_diagram.png')
```

```python
file_path = 'compact_phase_diagram3.txt'
m1 = np.genfromtxt(file_path, delimiter=' ')
```

```python
phase_diagram(m1,0.05,0.05)
```
