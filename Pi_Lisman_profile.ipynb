```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch
```

```python
m = np.zeros((80, 45))


m[0:4] = np.array([0]*5+[0]*5+[0]*5+[0]*5+[0]*10+[0]*5+[0]*5+[0]*5) #0.05-0.2s
m[4]   = np.array([0]*5+[0]*5+[0]*5+[0]*5+[0]*10+[1]*5+[1]*5+[0]*5) #0.25s 
m[5:13] = np.array([0]*5+[0]*5+[0]*5+[0]*5+[0]*10+[1]*5+[1]*5+[1]*5) #0.3-0.65s
m[13] = np.array([0]*5+[0]*5+[0]*5+[-1]*5+[-1]*4+[0]*5+[1]+[1]*5+[1]*5+[1]*5) #0.7s
m[13:21] = np.array([0]*5+[0]*5+[-1]*5+[-1]*5+[-1]*3+[0]*6+[1]+[1]*5+[1]*5+[1]*5)
m[21:37] = np.array([0]*5+[0]*5+[-1]*5+[-1]*5+[-1]*4+[0]*5+[1]+[1]*5+[1]*5+[1]*5) #0.75-1.80s
m[37:80] = np.array([0]*5+[-1]*5+[-1]*5+[-1]*5+[-1]*4+[0]*5+[1]+[1]*5+[1]*5+[1]*5) #1.85-2s
```

```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch

# Create a colormap based on your specifications
colors = ['white', 'blue', 'red']  # Order is important here
cmap = ListedColormap(colors)

# Convert matrix values to match the colormap
m[m == 0] = 0   # Mapping 0 to white
m[m == 1] = 1   # Mapping 1 to blue
m[m == -1] = 2  # Mapping -1 to red

# Display the matrix
fig, ax = plt.subplots(figsize=(6, 5))  # Adjusted to a smaller figure size
ax.imshow(m, cmap=cmap, aspect='auto', origin='lower')

# Display grid/lattice for the entire 80x45 matrix
ax.set_xticks(np.arange(-0.5, 45, 1), minor=True)
ax.set_yticks(np.arange(-0.5, 80, 1), minor=True)
ax.grid(which='minor', color='black', linestyle='-', linewidth=1, alpha=0.1)

# Set custom tick labels for y-axis (T(seconds))
y_ticks = np.arange(0, 80, 10)
y_labels = ["{:.1f}".format(0.05 * i) for i in range(10, 81, 10)]
ax.set_yticks(y_ticks)
ax.set_yticklabels(y_labels)
ax.set_ylabel('T(seconds)')

# Set custom tick labels for x-axis (Calcium Impulse(μm))
x_ticks = np.arange(4, 45, 5)
x_labels = ["{:.1f}".format(0.1 * i) for i in range(5, 46, 5)]
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_labels)
ax.set_xlabel('Calcium Impulse(μm)')

# Create legend
legend_labels = {
    'red': 'LTD',
    'blue': 'LTP',
    'white': 'basal'
}
handles = [Patch(facecolor=color, edgecolor='black', label=label) for color, label in legend_labels.items()]
ax.legend(handles=handles, loc='upper right')

plt.tight_layout()
plt.show()






```
